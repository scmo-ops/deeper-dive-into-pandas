```python
# Dependencies and Setup

import pandas as pd

# File to Load

df_path = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame

purchase_data = pd.read_csv(df_path)
```

## Player Count

* Display the total number of players


```python
# First, I thought the lenght of the list would be enough,

player_count=len(purchase_data)

# but then I wondered if some users purchased multiple items:

player_count = purchase_data['SN'].unique()
player_count1=len(player_count)
player_count1

# So, I really dont undestand whether the total number of players is 780 since the total number of unique players on the platform is 576,
# but some bought multiple items. Though for the next exersices I needed multiple info of the unique SN's.

player_count2 = purchase_data.loc[:, ['Gender', 'SN', 'Age']]
player_count2 = player_count2.drop_duplicates()
player_count3=len(player_count2)
player_count3
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculatios

unique_items = purchase_data['Item Name'].unique()
ui = len(unique_items)
average_price = purchase_data['Price'].mean()
gains = purchase_data['Price'].sum()


average_age = player_count2['Age'].mean()
std_age = player_count2['Age'].std()



# The summary data frame

summary_df = pd.DataFrame({'Unique Items' : [ui],'Average Price': [average_price],'Gains' : [gains],'Average age':[average_age],'Age std':[std_age]})

# Format adjustments for the data frame

summary_df['Average Price'] = summary_df['Average Price'].map("${:,.2f}".format)
summary_df['Gains'] = summary_df['Gains'].map("${:,.2f}".format)
summary_df['Average age'] = summary_df['Average age'].map("{:,.0f}".format)

#Final summary df

summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# First I need to group the data by the genders and then count them

gender_count = purchase_data.groupby(['Gender']).count()['Purchase ID'] # Since any column would work to count the values I chose
# ID value, but for no particular reason.

# Now I'll make the data group given the values of the distinct genders, and take the average price of those genders

average_price_gender = purchase_data.groupby(['Gender']).mean()['Price']

# This one is to get the gains of the different genders
gains_gender= purchase_data.groupby(['Gender']).sum()['Price']

# I thinik std is important when looking at data frames since one can see how centrated is the data.

std_gender= purchase_data.groupby(['Gender']).std()['Price']

# Summary table 

gender_analysis = pd.DataFrame({'Purchase count': gender_count, 'Average pricing': average_price_gender, 'Gains': gains_gender, 'Std of total pricing': std_gender})

# Add the format to the data

gender_analysis['Purchase count']= gender_analysis['Purchase count'].map('{:,}'.format)
gender_analysis['Average pricing']= gender_analysis['Average pricing'].map('${:,.2f}'.format)
gender_analysis['Gains']= gender_analysis['Gains'].map('${:,.2f}'.format)
gender_analysis['Std of total pricing']= gender_analysis['Std of total pricing'].map('{:,.2f}'.format)

# Display the table

gender_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Bins for the ages

age_bins = [0. , 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 200.0]
age_categories = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorization 

player_count2['Age bins'] = pd.cut( player_count2['Age'], age_bins, labels=age_categories)

# Calculations for the count and percentage

bin_count = player_count2['Age bins'].value_counts()
bin_percentage = (bin_count/ player_count3) *100

# Data frame and format

age_demographics = pd.DataFrame({"Total Count": bin_count, "Percentage of Players": bin_percentage})
age_demographics['Percentage of Players'] = age_demographics['Percentage of Players'].map('{:,.2f}%'.format)                  
age_demographics.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Binning and categorizing

age_bins = [0. , 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 200.0]
age_categories = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data['Age bins'] = pd.cut( purchase_data['Age'], age_bins, labels=age_categories)

# Calculations

group_count = purchase_data.groupby(['Age bins']).count()['Price']
average_pricing_age = purchase_data.groupby(['Age bins']).mean()['Price']
purchase_total_age = purchase_data.groupby(['Age bins']).sum()['Price']

std_person = purchase_data.groupby(['Age bins']).std()['Price']

# Data Frame

purchasing_analysis_age = pd.DataFrame({'Purchase count': group_count, 'Average purchase price': average_pricing_age, 'Total Purchase Value': purchase_total_age, 'Std of purchases': std_person  })

# Format

purchasing_analysis_age['Average purchase price'] = purchasing_analysis_age['Average purchase price'].map('${:,.2f}'.format)
purchasing_analysis_age['Total Purchase Value'] = purchasing_analysis_age['Total Purchase Value'].map('${:,.2f}'.format)
purchasing_analysis_age['Std of purchases'] = purchasing_analysis_age['Std of purchases'].map('{:,.2f}'.format)

# Display table

purchasing_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Calculations fot top spenders

purchase_count = purchase_data.groupby(['SN']).count()['Price']
average_price = purchase_data.groupby(['SN']).mean()['Price']
total_value = purchase_data.groupby(['SN']).sum()['Price']

# Data frame

top_spenders = pd.DataFrame({'Purchase Count': purchase_count, 'Average Purchase Price': average_price, 'Total Purchase Value': total_value})

# Format 
top_spenders['Average Purchase Price']=top_spenders['Average Purchase Price'].map('${:,.2f}'.format)

# Display table in descending order

top_spenders.sort_values("Total Purchase Value", ascending=False).head(19)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Calculations for the items

popular_items_count = purchase_data.groupby(['Item ID', 'Item Name']).count()['Price']
popular_items_price = purchase_data.groupby(['Item ID', 'Item Name']).mean()['Price']
popular_items_value = purchase_data.groupby(['Item ID', 'Item Name']).sum()['Price']

# Data frame

popular_items=pd.DataFrame({'Purchase Count': popular_items_count, 'Item Price': popular_items_price, 'Total Purchase Value': popular_items_value })

#Formating

popular_items['Item Price'] = popular_items['Item Price'].map('${:,.2f}'.format)
popular_items['Total Purchase Value'] = popular_items['Total Purchase Value'].map('${:,.2f}'.format) 

# Display the table
popular_items.sort_values('Purchase Count', ascending=False).head(15)
```
